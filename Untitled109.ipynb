### LOADING OF DATASET

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/HP/Downloads/Nvidia_stock_data.csv")
```

### DATA CLEANING

```python
# covert date to datetime and set as index
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.set_index('Date')
df = df.sort_index()
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.duplicated()
```

```python
# Fix column order
df = df[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
df.head()
```

```python
cols = ['Open', 'High', 'Low', 'Close', 'Volume']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

### Exploratory Data Analysis

```python
df.info()
```

```python
df.describe()
```

```python
# Date range
print("\n Date Range:", df.index.min(), "to", df.index.max())
```

```python
# Correlation matrix
print("\n Correlation:\n", df.corr())
```

```python
# Histogram of closing prices
df['Close'].plot(kind='hist', bins=50, figsize=(10, 4), title="Distribution of Close Prices")
plt.xlabel("Close Price")
plt.show()
```

```python
df['Close'].plot(figsize=(14, 5), title='NVIDIA Closing Prices Over Time')
plt.ylabel("Price ($)")
plt.show()
```

```python
# Volume over time
df['Volume'].plot(figsize=(14, 4), color='orange', title='Volume Traded Over Time')
plt.ylabel("Volume")
plt.show()
```

```python
#Highest and lowest Prices
print("Highest Close:\n", df['Close'].idxmax(), df['Close'].max())
print("Lowest Close:\n", df['Close'].idxmin(), df['Close'].min())
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df['Close'], label='Close')
plt.axhline(df['Close'].max(), color='green', linestyle='--', label='Max')
plt.axhline(df['Close'].min(), color='red', linestyle='--', label='Min')
plt.title('Closing Price with Max & Min')
plt.legend()
plt.show()
```

```python
# Moving Averages(7 Day &30 Day)
df['MA_7'] = df['Close'].rolling(7).mean()
df['MA_30'] = df['Close'].rolling(30).mean()
print(df[['Close', 'MA_7', 'MA_30']].tail())
```

```python
df[['Close', 'MA_7', 'MA_30']].plot(figsize=(14, 5), title="Moving Averages")
plt.ylabel("Price ($)")
plt.show()
```

```python
#Daiy Returns
df['Daily_Return'] = df['Close'].pct_change()
print(df['Daily_Return'].describe())
```

```python
df['Daily_Return'].hist(bins=100, figsize=(10, 4))
plt.title("Histogram of Daily Returns")
plt.xlabel("Return")
plt.show()
```

```python
#Volatility (30-Day Rolling Std)
df['Volatility_30'] = df['Close'].rolling(30).std()
print(df[['Close', 'Volatility_30']].dropna().tail())
df['Volatility_30'].plot(figsize=(14, 4), color='orange', title='30-Day Rolling Volatility')
plt.ylabel('Std Dev')
plt.show()
```

```python
#Monthly Average Close
monthly_avg = df['Close'].resample('M').mean()
print(monthly_avg.tail())
monthly_avg.plot(figsize=(14, 4), title="Monthly Avg Closing Price")
plt.ylabel("Price")
plt.show()
```

```python
#Yearly Average Close
yearly_avg = df['Close'].resample('Y').mean()
print(yearly_avg)
yearly_avg.plot(figsize=(12, 4), title="Yearly Avg Close")
plt.ylabel("Price")
plt.show()
```

```python
df['Year'] = df.index.year
summary_by_year = df.groupby('Year').agg({
    'Close': ['mean', 'max', 'min'],
    'Volume': 'mean'
})
print(summary_by_year)
```

```python
correlation = df[['Open', 'High', 'Low', 'Close', 'Volume']].corr()
print(correlation)
```

```python
plt.matshow(correlation, cmap='coolwarm')
plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=90)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.colorbar()
plt.title("Correlation Heatmap", pad=20)
plt.show()
```

```python
#Top 10 Volume days
top_volume = df[['Close', 'Volume']].sort_values(by='Volume', ascending=False).head(10)
print(top_volume)
```

```python
#Monthly Avg Pivot Table (Year vs Month)
df['Month'] = df.index.month
pivot = df.pivot_table(values='Close', index='Year', columns='Month', aggfunc='mean')
print(pivot.round(2))
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(pivot, aspect='auto', cmap='YlGnBu')
plt.title("Monthly Avg Close (Heat View)")
plt.xlabel("Month")
plt.ylabel("Year")
plt.colorbar(label="Price")
plt.show()
```
